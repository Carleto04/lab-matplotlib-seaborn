#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# MODO1
x = np.arange(0,100)
y1 = x**2
y2 = np.exp(x)

fig, ax = plt.subplots(1, 2, figsize=(10,5))
ax[0].plot(x, y1)
ax[1].plot(x, y2)

ax[0].set_xscale('linear') #what values will be in x axis
ax[0].set_yscale('linear') #what values will be in y axis

ax[1].set_xscale('log') #what values will be in x axis
ax[1].set_yscale('log') #what values will be in y axis
```

```python
# MODO2
x = np.arange(0,100)
y1 = x**2
y2 = np.exp(x)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))
ax1.plot(x, y1)
ax2.plot(x, y2)

ax1.set_xscale('linear') #what values will be in x axis
ax1.set_yscale('linear') #what values will be in y axis

ax2.set_xscale('log') #what values will be in x axis
ax2.set_yscale('log') #what values will be in y axis
```
